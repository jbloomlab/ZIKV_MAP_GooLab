## Analyze C8 neutralization data 
We generated single mutant reporter virus particles (RVPs) based on DMS differential selection analysis. These RVPs have a GFP based readout that we analyzed via high throughput flow cytometry. Now, we want to determine if single mutations contributed to antibody neutralization resistance or antibody escape. We also want to make figures for the paper. 

```python
# Import packages
import pandas as pd
import os
import numpy as np
from scipy.stats import ttest_ind
from IPython.display import HTML, display
from statsmodels.sandbox.stats.multicomp import multipletests
import operator

import neutcurve
```

```python
# Set pandas display options
pd.set_option('display.float_format', '{:.4g}'.format)
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 500)
```

```python
## ID sequences and locations
datadir = './data'
resultsdir ='./results'
os.makedirs(datadir, exist_ok = True)
os.makedirs(resultsdir, exist_ok = True)

# ID antibody
antibody = 'C8'

# ID sequence
platereader_datafile = os.path.join(datadir,'C8_CK061_neutdata.csv')
platereader = pd.read_csv(platereader_datafile)
```

### Calculate fraction infectivity
Based on Kate Crawford's SARS-CoV-2 JID paper, we get calculate fraction infectivity by dividing individual datapoints by the average of no-antibody or no-serum controls.

```python
# Get list of unique viruses and sort them by the mutated site
viruses = []
for item in platereader.columns:
    item = item.split('_')[0]
    if item != antibody:
        if item == 'WT':
            site = 0
        else:
            site = int(item[1:-1])
        viruses.append([item, site])
viruses = sorted(list(set(tuple(row) for row in viruses)), key=lambda x: x[1])
viruses = [i[0] for i in viruses]

virus_index = dict(zip(viruses, range(len(viruses))))

# Get list of concentrations
concentrations = []
for item in (platereader[f'{antibody}']):
    if item != 'no_ab':
        concentrations.append(item)

# Get no_ab average to utilize as denominator for fraction infectivity calculation
# Then divide each value by this denominator
# Save to reformatted dataframe to be parsed by neutcurve
neutcurve_ls = []
for v in viruses:
    rep1 = v + '_1'
    rep2 = v + '_2'
    rep3 = v + '_3'
    replicates = (rep1, rep2, rep3)
    no_ab_average = (platereader[[rep1,rep2,rep3]].loc[platereader[antibody] == 'no_ab'].sum(axis=1)/3).values[0]    
    
    for i in range(3):
        rep = i+1
        
        for n in range(0, len(concentrations)):
            conc = concentrations[n]
            frac_infectivity = platereader[f'{v}_{rep}'][n] / no_ab_average
            
            if frac_infectivity == 0:
                frac_infectivity = 0.00000001
            
            neutcurve_ls.append([antibody, v, rep, conc, float(frac_infectivity)])
            
            
# Get dataframe with columns 'serum','virus','replicate','concentration','fraction infectivity'
# neutcurve is set up to accept only columns with these names
neutcurve_df = (pd.DataFrame(data=np.array(neutcurve_ls),
    columns=['serum','virus','replicate','concentration','fraction infectivity'])
                .astype({'fraction infectivity': 'float64', 'concentration': 'float64'})
               )
```

```python
neutcurve_df 
```

### Fit Hill curves to data with neutcurve
Using the neutcurve documentation, we can produce replicate Hill curves and IC50s for our datasets. With the tables of IC50s, we can perform statistical tests. We can also plot these curves for figures for the paper. 

```python
# Call CurveFits function from neutcurve
fits = neutcurve.CurveFits(neutcurve_df)
fits.replicates
```

```python
# Basic plots
fig, axes = fits.plotSera(xlabel = 'concentration (ug/ml)')
```

```python
# Get replicate and average IC50 values
fits_replicates = (
    fits.fitParams(average_only=False)
    .pivot(index = 'virus', columns = 'replicate', values = 'ic50')
    .reset_index()
    .rename(columns = {'1': 'replicate-1', '2': 'replicate-2',
                       '3': 'replicate-3', 'average': 'mean'})
    .rename_axis(None, axis=1)
    )


# Calculate p-values with unpaired students t-test
WT_IC50 = fits_replicates.query('virus == "WT"')[['replicate-1','replicate-2','replicate-3']].values.flatten()
p_values = []
for v in fits_replicates['virus']: 
    # Some of our curves seem drastically different than other 2 fits...
    if v == 'T315K' or v == 'T315C' or v == 'K373H':
        mutant_IC50 = fits_replicates.query(f'virus == "{v}"')[['replicate-1','replicate-3']].values.flatten()
    elif v == 'T315Y':
        mutant_IC50 = fits_replicates.query(f'virus == "{v}"')[['replicate-1','replicate-2']].values.flatten()
    else:
        mutant_IC50 = fits_replicates.query(f'virus == "{v}"')[['replicate-1','replicate-2','replicate-3']].values.flatten()
                
    p_val = ttest_ind(WT_IC50, mutant_IC50, equal_var=False)
    p_values.append(p_val[1])

# Bonferroni correction 
alpha = 0.05 # Level of significance to return bool
bonferroni_p_values = multipletests(p_values, alpha=alpha, method='bonferroni', is_sorted=False)[1]
ben_hoch_fdr_p_values = multipletests(p_values, alpha=alpha, method='fdr_bh', is_sorted=False)[1]

# Add p-values to dataframe
fits_replicates['p_val'] = p_values
fits_replicates['bonferroni_p_val'] = bonferroni_p_values
fits_replicates['benhochfdr_p_val'] = ben_hoch_fdr_p_values
    
# Reindex based on virus order (WT first, then by increasing mutated site index)
# params_avg_reps['virus_rank'] = params_avg_reps['virus'].map(virus_index)
# params_avg_reps.sort_values(['virus_rank'],
#         ascending = [True], inplace = True)
# params_avg_reps.drop('virus_rank', axis = 1, inplace = True)

fits_replicates
```

### Conclusions
It seems like our intrareplicate variability is so large that Bonferroni corrections potentially overinflate p-values. Not sure what is rigorous but still appears to actually reflect level of difference between WT and mutant virus neutralization curves. I added Benjamini-Hochberg FDR corrected p-value correction, which is less stringent at the risk of more false positives, relative to Bonferroni.
