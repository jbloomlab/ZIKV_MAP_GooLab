## Analyze all single mutant neutralization data 
We generated single mutant reporter virus particles (RVPs) based on DMS differential selection analysis. These RVPs have a GFP based readout that we analyzed via high throughput flow cytometry. Now, we want to determine if single mutations contributed to antibody neutralization resistance or antibody escape. We also want to make figures for the paper. 

```python
# Import packages
import pandas as pd
import os
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
import scipy.stats
import operator
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import neutcurve
```

```python
# Set pandas display options
pd.set_option('display.float_format', '{:.4g}'.format)
pd.set_option('display.max_columns', 20)
pd.set_option('display.width', 500)
```

```python
## ID sequences and locations
datadir = './data'
resultsdir ='./results'
os.makedirs(datadir, exist_ok = True)
os.makedirs(resultsdir, exist_ok = True)

# ID antibody and experiment number

ab_dict = {'MR766': {'C10': 'CC60',
              'C8': 'CC60',
              'SiGN3C': 'CC60',
              'MZ4': 'CC60',
              'ZV67': 'CK071'},
           
           'DENV2': {'C10': 'CK078',
              'C8': 'CK078',
              'SiGN3C': 'CK078',
              'MZ4': 'CK078',},
           
           'HPF2013': {'C10': 'CK079',
                'C8': 'CK079',
                'SiGN3C': 'CK079',
                'MZ4': 'CK079',}
          }
```

### Calculate fraction infectivity
Based on Kate Crawford's SARS-CoV-2 JID paper, we get calculate fraction infectivity by dividing individual datapoints by the average of no-antibody or no-serum controls.

```python
neutcurve_ls = []

for virus in ab_dict:
    
    print(virus)

    for antibody in ab_dict[virus]:
        # ID sequence
        platereader_datafile = os.path.join(datadir,f'{antibody}_{ab_dict[virus][antibody]}_neutdata.csv')
        platereader = pd.read_csv(platereader_datafile)

        # Get list of unique viruses and sort them by the mutated site
        viruses = []
        for item in platereader.columns:
            item = item.split('_')[0]
            if item != antibody:
                if item == 'WT':
                    site = 0
                else:
                    site = int(item[1:-1])
                viruses.append([item, site])
        viruses = sorted(list(set(tuple(row) for row in viruses)), key=lambda x: x[1])
        viruses = [i[0] for i in viruses]

        # Get list of concentrations
        concentrations = []
        for item in (platereader[f'{antibody}']):
            if item != 'no_ab':
                concentrations.append(item)

        # Caclulate fraction infectivity and then append to neutcurve_ls
        # Get no_ab average to utilize as denominator for fraction infectivity calculation
        # Then divide each value by this denominator
        # Save to reformatted dataframe to be parsed by neutcurve
        for v in viruses:
            rep1 = v + '_1'
            rep2 = v + '_2'
            rep3 = v + '_3'
            replicates = (rep1, rep2, rep3)
            no_ab_average = (platereader[[rep1,rep2,rep3]].loc[platereader[antibody] == 'no_ab'].sum(axis=1)/3).values[0]    

            # Add replicate number
            for i in range(0, 3):
                rep = i+1

                # Add concentration label
                for n in range(0, len(concentrations)):
                    conc = concentrations[n]
                    frac_infectivity = platereader[f'{v}_{rep}'][n] / no_ab_average

                    # Neutcurve doesn't like 0's...?
                    if frac_infectivity == 0:
                        frac_infectivity = 0.00000001
                        
                    # Replace v with a virus label
                    virus_labeled = v + '_' + virus
                
                    neutcurve_ls.append([antibody, virus_labeled, rep, conc, float(frac_infectivity)])


# Get dataframe with columns 'serum','virus','replicate','concentration','fraction infectivity'
# neutcurve is set up to accept only columns with these names
neutcurve_df = (pd.DataFrame(data=np.array(neutcurve_ls),
    columns=['serum','virus','replicate','concentration','fraction infectivity'])
                .astype({'fraction infectivity': 'float64', 'concentration': 'float64'})
               )

neutcurve_df['serum'] = (neutcurve_df['serum']
#                          .str.replace('SiGN3C', 'SIgN3C')
                        )
```

### Fit Hill curves to data with neutcurve
Using the neutcurve documentation, we can produce replicate Hill curves and IC50s for our datasets. With the tables of IC50s, we can perform statistical tests. We can also plot these curves for figures for the paper. 

```python
# Call CurveFits function from neutcurve
fits = neutcurve.CurveFits(neutcurve_df)
```

```python
# Color palette
colors = {'teal': '#3bceac',
          'red': '#DA2C38',
          'blue': '#446df6',
          'yellow': '#F6AE2D',
          'green': '#00635D',
         }

colorwheel = colors.values()

# Markers
m = 'o'
```

```python
# Utility function to calculate f test to determine if equal variance assumption can be made
def f_test(group1, group2):
    x = np.array(group1)
    y = np.array(group2)
    f = np.var(group1, ddof=1)/np.var(group2, ddof=1)
    nun = x.size-1
    dun = y.size-1
    p_value = scipy.stats.f.cdf(f, nun, dun)
    return p_value

# Define function to retrieve variety of stats for all viruses (across all neuts) specified by list
def get_stats(data, virus_lst, outfile):
    
    # Initialize empty list for p-values
    p_values = []
    
    
    # Interation through viruses 
    for background in ab_dict:
      
        # Subset the subsetted dataframe on virus background
        subset_df = data.query(f'background == "{background}"')
        
        
        # Iteration through antibodies
        for antibody in ab_dict[background]:
            
            # Subset the subsetted dataframe on antibody
            resubset_df = subset_df.query(f'serum == "{antibody}"')
                          
            # Get wildtype IC50s to compare to
            WT_IC50 = resubset_df.query('mutation == "WT"')[['rep1-IC50','rep2-IC50','rep3-IC50']].values.flatten()
                        
            # Then iterate through each virus to get mut IC50s
            for v in resubset_df['mutation']:
                # Get stats for difference between mutant and WT IC50s
                if v !='WT':
                    mutant_IC50 = (resubset_df.query(f'mutation == "{v}"')
                                   [['rep1-IC50','rep2-IC50','rep3-IC50']]
                                   .values.flatten())
        
                    # Calculate fold-change from WT
                    foldchange = mutant_IC50/WT_IC50
                    # Calcualte mean fold-change
                    logmeanFC = np.log(np.mean(foldchange))
                    # Calculate p-values with unpaired students t-test
                    ttest = (ttest_ind(WT_IC50, mutant_IC50, equal_var=False)[1])
                    # Calculate f-test to determine in variance is equal between datasets
                    ftest = f_test(WT_IC50, mutant_IC50)
                    # Calculate p-values with 1-way ANOVA
                    ANOVA = (f_oneway(WT_IC50, mutant_IC50)[1])
                
                # Comparison of WT to WT should yield none-type
                elif v == 'WT':
                    foldchange = (np.nan)
                    meanfoldchange = (np.nan)
                    ttest = (np.nan)
                    ftest = (np.nan)
                    ANOVA = (np.nan)      
                
                p_values.append([background, antibody, v, foldchange, logmeanFC, ttest, ftest, ANOVA])
                

    # Create p-value dataframe and merge with replicate IC50 data
    pval_df = pd.DataFrame(p_values, columns=['background',
                                              'serum',
                                              'mutation', 
                                              'foldchange', 
                                              'logmeanFC',
                                              'ttest', 
                                              'ftest',
                                              'ANOVA'])
    
    data = data.merge(pval_df, on=['background', 'serum','mutation'])
    
    # fits_subset.to_latex(buf = os.path.join(resultsdir, 'all_abs_IC50.tex'), index=False)
    data.to_csv(os.path.join(resultsdir, outfile), index=False)
    return(data)
```

```python
# Call function for all viruses
fits_all = (
    fits.fitParams(average_only=False)
    .pivot(index = ['serum','virus'], columns = 'replicate', values = 'ic50')
    .reset_index()
    .rename(columns = {'1': 'rep1-IC50', 
                       '2': 'rep2-IC50',
                       '3': 'rep3-IC50'})
    .rename_axis(None, axis=1)
    )

# Get average IC50s and the standard error accompanying for n=3
params_avg_reps = (
    fits.fitParams(average_only=False)
    .query('replicate != "average"')
    .groupby(['serum', 'virus'])
    .aggregate(ic50_replicate_avg=pd.NamedAgg('ic50', 'mean'),
               ic50_replicate_stderr=pd.NamedAgg('ic50', 'sem'),
               )
    .reset_index()
    )


fits_all = (fits_all
               .merge(params_avg_reps, on=['serum', 'virus'])
               .rename(columns = {'ic50_replicate_avg': 'mean-IC50',
                                  'ic50_replicate_stderr': 'stderr'})
               .assign(background = lambda x: x.virus.str.split('_').str[1],
                       mutation = lambda x: x.virus.str.split('_').str[0])
               .query('mutation != "71D"')
               .sort_values(by = ['serum', 'mutation', 'background'])
               [['serum','mutation','virus','rep1-IC50','rep2-IC50','rep3-IC50',
                 'average', 'mean-IC50', 'stderr', 'background']]
              )


fits_all.query('background == "DENV2"').head(20)

virus_lst = fits_all['mutation'].unique()



bigframe = get_stats(fits_all, virus_lst, outfile='all_nabs_fold-change.csv')

bigframe['serum'] = (bigframe['serum']
                     .str.replace('C10', 'EDE1-C10')
                     .str.replace('C8', 'EDE1-C8')
                     .str.replace('SiGN3C', 'SIgN-3C')
                     .str.replace('ZV67', 'ZV-67')
                    )
```

```python
virus_list = ['S7P', 'T49D', 'S72G']

for virus in bigframe.mutation.unique():
    
    if virus  in virus_list:

        temp_df = (bigframe
                   .query(f'mutation == "{virus}"')
                   .sort_values(by = ['serum', 'background'], ascending = [True, False])
                   .pivot('serum', 'background', 'logmeanFC')
                   [['MR766', 'HPF2013', 'DENV2']]
                  )

        print(temp_df)
        
        # Make name that ignores 'wt' aa since this will vary across backgrounds
        virusname = virus[1:]
        
        # Reset plot
        plt.figure()

        ax = sns.heatmap(temp_df, annot=True, 
                         cmap='RdBu_r', center = 0.0,
                         xticklabels = ['ZIKV-MR766', 'ZIKV-H/PF/2013', 'DENV2-16681'])
        ax.set(title = f'mutation = {virusname}', xlabel = '', ylabel = '')
        
        outfile = os.path.join(resultsdir, f'{virus}_heatmap.pdf')
        fig = ax.get_figure()
        fig.savefig(outfile, dpi = 'figure')
        
```

```python
# refine to just MR766 
mr766_df = bigframe.query('background == "MR766"')

# initialize empty dataframe
sera_maxFC_df = pd.DataFrame(columns = ['serum', 'mutation', 'foldchange', 'logmeanFC'])


for sera in mr766_df.serum.unique():
    # identify serum-specific values and local maxima
    sera_df = mr766_df.query(f'serum == "{sera}"')[['serum', 'mutation', 'foldchange', 'logmeanFC']]
    sera_max_index = sera_df['logmeanFC'].idxmax()
    
    # extract individual replicate FC values    
    temp_df = sera_df.loc[[sera_max_index]]
    temp_df['fc_rep1'] = (temp_df['foldchange'].values[0][0])
    temp_df['fc_rep2'] = (temp_df['foldchange'].values[0][1])
    temp_df['fc_rep3'] = (temp_df['foldchange'].values[0][2])
    
    # add max data FC reps and log-mean to new dataframe
    sera_maxFC_df = (sera_maxFC_df
                     .append(temp_df)
                     [['serum', 'mutation','foldchange', 'fc_rep1', 'fc_rep2', 'fc_rep3', 'logmeanFC']]
                    )
    
    
sera_maxFC_df = (sera_maxFC_df
                 .reset_index(drop=True)
                 .query('serum != "ZKA64"')
                )

sera_maxFC_df = pd.melt(sera_maxFC_df, id_vars=['serum', 'mutation'], value_vars=['fc_rep1', 'fc_rep2', 'fc_rep3'],
                       var_name='replicate', value_name = 'fc')



# Sort based on custom list
sorter = ['SIgN-3C', 'EDE1-C8', 'EDE1-C10', 'MZ4', 'ZV-67']
# Custom lexicographical sort
sorterIndex = dict(zip(sorter, range(len(sorter))))
sera_maxFC_df['antibody_rank'] = sera_maxFC_df['serum'].map(sorterIndex)
sera_maxFC_df.sort_values(['antibody_rank'],
        ascending = [False], inplace = True)
sera_maxFC_df = sera_maxFC_df.reset_index(drop=True)
```

```python
sera_maxFC_df['logfc'] = np.log(sera_maxFC_df['fc'])
```

```python
# load the dataset

p = sns.stripplot(data=sera_maxFC_df, x='serum', y='logfc', size=7, palette = colorwheel)

# plot the mean line
sns.boxplot(showmeans=True,
            meanline=True,
            meanprops={'color': 'k', 'ls': '-', 'lw': 2},
            medianprops={'visible': False},
            whiskerprops={'visible': False},
            zorder=10,
            x='serum', y='logfc',
            data=sera_maxFC_df,
            showfliers=False,
            showbox=False,
            showcaps=False,
            ax=p)

plt.ylabel('log fold-change IC50')
plt.xlabel('')
plt.show()


outfile = os.path.join(resultsdir, 'maxIC50shift_swarmplot.pdf')
fig = p.get_figure()
fig.savefig(outfile, dpi = 'figure')
```

```python
for antibody in tidy_df['serum'].unique():
    
    df = tidy_df.query(f'serum == "{antibody}"')

    if antibody == 'ZV-67':
        fig = plt.figure(figsize=(12,3), dpi=300)
        ax = plt.subplot()
        
    else:
        fig = plt.figure(figsize=(12,10), dpi=300)
        ax = plt.subplot()

    ncols = df.shape[1]
    nrows = df.shape[0]

    ax.set_xlim(0, ncols + 1)
    ax.set_ylim(0, nrows + 1)

    positions = [0.6, 1.8, 2.8, 3.8, 4.8, 5.8, 6.8, 8, 9.0]
    columns = ['serum', 'mutation', 'background', 'rep1-IC50', 'rep2-IC50', 'rep3-IC50', 'mean-IC50', 'stderr', 'ttest']

    # Add table's main text
    for i in range(nrows):
        for j, column in enumerate(columns):
            if j == 0:
                ha = 'left'
            else:
                ha = 'center'
            if 'IC50' in column:
                text_label = f'{df[column].iloc[i]:,.3f}'
            elif column == 'stderr':
                text_label = f'{df[column].iloc[i]:,.3f}'
            elif column == 'ttest':
                text_label = f'{df[column].iloc[i]:,.3f}'
                if df[column].iloc[i] < 0.05:
                    weight = 'bold'
                    style = 'oblique'
            else:
                text_label = f'{df[column].iloc[i]}'
                weight = 'normal'
                style = 'normal'
            ax.annotate(
                xy=(positions[j], i + .5),
                text=text_label,
                ha=ha,
                va='center',
                weight=weight,
                style=style
            )

    # Add column names
    column_names = ['serum', 'mutation', 'background', 'rep1-IC50', 'rep2-IC50', 'rep3-IC50', 'mean-IC50', 'stderr', 'ttest']
    for index, c in enumerate(column_names):
            if index == 0:
                ha = 'left'
            else:
                ha = 'center'
            ax.annotate(
                xy=(positions[index], nrows + .1),
                text=column_names[index],
                ha=ha,
                va='bottom',
                weight='bold'
            )

    # Add dividing lines
    ax.plot([ax.get_xlim()[0], ax.get_xlim()[1]], [nrows, nrows], lw=1.5, color='black', marker='', zorder=4)
    ax.plot([ax.get_xlim()[0], ax.get_xlim()[1]], [0, 0], lw=1.5, color='black', marker='', zorder=4)
    for x in range(1, nrows):
        ax.plot([ax.get_xlim()[0], ax.get_xlim()[1]], [x, x], lw=1.15, color='gray', ls=':', zorder=3 , marker='')

    ax.set_axis_off()
    plt.savefig(
        f'results/{antibody}_IC50_statistical_tests.png',
        dpi=300,
        transparent=True,
        bbox_inches='tight'
    )
```
